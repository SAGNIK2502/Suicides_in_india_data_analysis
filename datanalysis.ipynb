```python
Dat=d.groupby('Year').Total.sum()
Dat
```

```python
x=Dat.index.values.reshape(-1,1)
x
```

```python
y=Dat.values
y
```

```python
plt.scatter(x,y)
plt.show()
```

Polynomial regression

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=5)
X=poly.fit_transform(x)
pd.DataFrame((X))
```

```python
from sklearn.linear_model import LinearRegression
lin=LinearRegression()
lin.fit(X,y)
```

```python
lin.coef_
```

```python
lin.intercept_
```

```python
y
```

```python
yp=lin.predict(X)
yp
```

```python
plt.scatter(x,y)
plt.plot(x,yp)
plt.show()
```

```python
lin.score(X,y)*100
```
